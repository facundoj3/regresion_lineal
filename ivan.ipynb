```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Simular datos (reemplaza esto con tus datos reales)
# 'partido' es la variable independiente y 'IDLN105' es la dependiente
data = {
    'partido': np.random.randn(110),  # Reemplaza con tus datos
    'IDLN105': 16.0924 - 0.4462 * np.random.randn(110) + np.random.randn(110)  # Ejemplo de datos generados
}

# Crear un DataFrame
df = pd.DataFrame(data)

# Definir variable dependiente y la independiente
X = df['partido']
y = df['IDLN105']

# Añadir una constante a la variable independiente
X = sm.add_constant(X)

# Ajustar el modelo OLS
model = sm.OLS(y, X).fit()

# Obtener los resultados
print(model.summary())

# Hacer predicciones
predictions = model.predict(X)

# Graficar los datos y la línea de regresión
plt.figure(figsize=(10, 6))
plt.scatter(df['partido'], y, color='blue', label='Datos Observados')
plt.plot(df['partido'], predictions, color='red', label='Línea de Regresión', linewidth=2)
plt.title('SERGIO TOMAS')
plt.xlabel('Partido')
plt.ylabel('IDLN105')
plt.legend()
plt.grid()
plt.show()
```
